```python
import pandas as pd
import numpy as np
import glob
from tqdm import tqdm
import itertools
import os
```

```python
match_performance_list = glob.glob('../../data/result/tool_performance/*/*match_performance.tsv')
pas_quantification_performance_list = glob.glob('../../data/result/tool_performance/*/*pas_quantify_performance.tsv')
de_apa_performance_list = glob.glob('../../data/result/tool_performance/*/*de_apa_performance.tsv')

if not os.path.exists(
    '../../data/result/merged_tool_performance'
):
    os.makedirs(
        '../../data/result/merged_tool_performance'
)
```

```python
# 更简洁的写法：使用列表推导式（List Comprehension）
# 功能同上，代码更紧凑
match_performance_df = pd.concat(
    [pd.read_csv(f, sep='\t') for f in tqdm(match_performance_list)], 
    ignore_index=True
)
```

```python
match_performance_df.to_csv('../../data/result/merged_tool_performance/match_performance.tsv', sep='\t', index=False)
```

```python
pas_quantification_performance_df = pd.concat(
    [pd.read_csv(f, sep='\t') for f in tqdm(pas_quantification_performance_list)], 
    ignore_index=True
)
```

```python
pas_quantification_performance_df.to_csv('../../data/result/merged_tool_performance/pas_quantification_performance.tsv', sep='\t', index=False)
```

```python
de_apa_performance_df = pd.concat(
    [pd.read_csv(f, sep='\t') for f in tqdm(de_apa_performance_list)], 
    ignore_index=True
)
```

```python
de_apa_performance_df.to_feather('../../data/result/merged_tool_performance/de_apa_performance.feather')
```
